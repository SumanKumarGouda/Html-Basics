Question 1 : Explain the basic structure of an HTML page. Why is it important to follow this structure?
 ANS-Basic Structure of an html page:



```python

!DOCTYPE html>: Tells the browser that the document is an HTML5 page.

<html lang="en">: This is the root element of an HTML page. The lang="en" attribute specifies the language of the document, which helps with accessibility and search engine optimization.

<head>: #This section contains meta-information about the HTML document that is not displayed directly on the page. It includes:
<meta charset="UTF-8">: #Specifies the character encoding for the document, ensuring that text is displayed correctly.
<meta name="viewport" content="width=device-width, initial-scale=1.0">: #Configures the viewport for responsive web design, making the page look good on various devices.
</head>: #closing head tag.

<title>Document</title>:Sets the title of the HTML page, which appears in the browser's title bar or tab.

<body>: This section contains the visible content of the HTML document, such as text, images, links, and other elements that users interact with.
</body>: closing body tag.
</html>: closing html tag.


this structure is important because:
1. Browser understanding.
2. Consistency across devices.
3. SEO (Search Engine Optimization).
4. Accessibility.
5. Maintainability.
6. Required by HTML standards.
```

Question 2: You’re building a small personal profile page. Which HTML tags will you use to highlight your name as a heading, describe yourself in a paragraph, and emphasize a key skill?
  ANS-<h1> tag is used for highlighting my name as a heading,because it represents the most important heading on the page.

```python

<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>

<body>
<h1>Suman</h1>
#Paragraph to describe my self.
<p>My name is Suman,I am a passionate Fullstack Developer student. I love building web applications and solving complex problems with code.
</p>
#Emphasize a key skill.
<p>My primary expertise is in <strong>Java Programming,HTML,CSS</strong>.
</p>
</body>
</html>

```

Question 4: You’re creating a form that asks the user to enter their name, select their country from a list, and click a button to submit. Write the HTML code for this.
 ANS-

```python

<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
     <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Form</title>
  </head>
  <body>
    <h1>Survey Form</h1>
    <form>
      <label for="name">Enter Your Name:</label>
      <input type="text" name:"Name" id="name" placeholder="Enter your Name"></input>
      <br>
      <label for="country">Select Your Country:</label>
      <select name="country" id="country" placeholder="select your country">
       <option value="in">India</option>
       <option value="Us">USA</option>
       <option value="Uk">United Kindom</option>
      </select>
    </form>
  </body>
  </html>
```
